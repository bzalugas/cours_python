## Variables
Les variables sont un des éléments les plus récurrents des langages de programmation. Une variable est comme une boîte permettant de stocker une information : un chiffre, un texte, le résultat d'un calcul, et tous les types de données que nous verrons plus tard.
Il faut imaginer donc qu'une variable est une boîte et qu'elle possède une étiquette permettant de l'identifier. Lorsque nous enregistrons une donnée dans une variable, pour retrouver cette donnée il suffit d'utiliser le nom de la variable, c'est-à-dire son étiquette.

Dans l'exemple ci-dessous, nous créons une variable qui s'appelle `nombre`et qui contient le nombre **23**.
```python
nombre = 23
```
Nous pouvons afficher le contenu de la variable en utilisant la fonction `print` et le nom de la variable : 
```python
print(nombre)
```
On peut également effectuer des calculs avec le contenu des variables :
```python

resultat = nombre + 5
print(resultat)
```
Dans cet exemple, 28 sera affiché dans la console.

# Les conditions
Jusqu'ici, nous avons donné une suite d'instructions à l'ordinateur pour qu'il les fasse toutes une par une sans exception. Cela suffit pour les programmes très simples mais nous allons vite avoir besoin de pouvoir exécuter certaines instructions seulement **dans certaines conditions**.

Pour exécuter une instruction en fonction d'une condition, on utilise le mot-clef `if` suivi de la **condition** et du caractère ":" en fin de ligne. On écrit ensuite la ou les instructions à effectuer en sautant une ligne et avec une **indentation**.

**Exemple** : 
```python
age = 16
if age < 18:
    print("L'âge est inférieur à 18.")
    print("Cela signifie que tu n'es pas encore majeur.")
print("J'exécute cette ligne.")
```
Le code ci-dessus va afficher tous les textes car 16 est inférieur à 18. La ligne qui affiche "J'exécute cette ligne" va également s'afficher dans tous les cas, peut importe la valeur de la variable `age` car il n'y a pas d'**indentation** au début de la ligne, ce qui signifie qu'elle ne dépend pas d'une condition ou d'un autre type de bloc.

# If, else : si-alors-sinon
L'instruction `if`, que nous pourrions traduire par "si, alors", fait quelque chose lorsqu'une condition est vraie (`True`). Dans certains cas, on veut prévoir en plus de faire autre chose quand la conditition n'est pas vraie. Nous pourrions par exemple afficher un message à l'écran si la variable `age` vaut 18 et un autre message si `age` ne vaut pas 18 (`False`).

**Exemple** :
```python
age = 16
if age == 18:
    print("L'âge est égal à 18.")
else:
    print("L'âge est différent de 18.")
print("J'exécute cette ligne.")
```
Copie-colle cet exemple en changeant la valeur de `age` pour comprendre comment fonctionnent les instructions `if` et `else`.

````python
age = 16
if age == 18:
    print("L'âge est égal à 18.")
else:
    print("L'âge est différent de 18.")
print("J'exécute cette ligne.")
````

# if et elif
Nous pouvons même aller plus loin avec le mot-clef `elif`, qui est l'abréviation de **else-if**, **"sinon-si"**. Par exemple, si nous voulons vérifier que l'âge d'une personne vaut 18, 20, 21 et ainsi de suite, pour que le programme fasse des choses différentes selon l'âge donné en réponse. Ces instructions sont différentes des instructions `if-else` car il peut y avoir **plusieurs `elif` dans la même instruction :**
```python
age = 18
if age == 18:
    print("Tu as 18 ans :")
    print("Tu as pile la majorité !")
elif age == 20:
    print("Bienvenue dans la vingtaine !")
    print("Plus qu'un an avant la majorité internationale !")
elif age == 21:
    print("21 ans, la majorité internationale !")
else:
    print("Je ne sais pas quel âge tu as.")
```

Dans cet exemple, les instructions d'une seule condition seront exécutées à chaque fois. Même s'il est parfois possible que plusieurs conditions soient vraies, seule la première à être vraie sera considérée et ses instructions exécutées.

# Les boucles
Les boucles permettent d'exécuter du code répétitif sans écrire plusieurs fois les mêmes instructions. Par exemple, si on veut afficher 5 fois le texte "Bonjour", on pourrait écrire ce code:
```python
print("Bonjour")
print("Bonjour")
print("Bonjour")
print("Bonjour")
print("Bonjour")
```
Cela prend donc 5 lignes et si finalement nous avions besoin de l'afficher 250 fois, cela prendrais 250 lignes. Voici l'équivalent de l'exemple précédent mais avec une boucle `for` (nous verrons plus tard les détails de la syntaxe) :
```python
for i in range(0, 5):
    print("Bonjour")
```

Nous avons atteint le même résultat en seulement deux lignes ! De plus, une répétition de 250 fois prendrait exactement le même nombre de lignes.

En python, il existe **2 types de boucles** : les boucles `for` et les boucles `while`. Nous consacrerons un chapitre par type de boucles.

## Les boucles for
Une boucle `for` permet d'éxecuter des instructions un nombre définit de fois. La syntaxe de ce type de boucle équivaut à écrire :

*POUR* une variable **nom_de_variable** prenant ses valeurs dans un ensemble, exécuter ces instructions :

    instruction1
    instruction2
    ...

Lorsque l'on veut répéter les instructions un nombre définit de fois, on utilise la fonction `range(start, end)`. Cette fonction nous renvoie l'ensemble de tous les nombres en commençant par `start` et en s'arrêtant à `end - 1`.

La syntaxe réelle est la suivante, en supposant que nos instructions soient d'afficher du texte :

```python
for ma_variable in range(x, y):
    print("Ligne 1")
    print("Ligne 2")
```

La variable `ma_variable` va prendre successivement toutes les valeurs allant de x à y **exclus**. Par exemple, **si x = 0 et y = 10**, la variable `ma_variable` aura d'abord la valeur *0*, les deux lignes vont s'afficher, puis la variable prendra la valeur *1*, et ainsi de suite jusqu'à ce que la variable contienne la valeur *9*. Après avoir affiché les lignes, la boucle s'arrêtera.

On peut ajouter un 3^{ème} paramètre à la fonction `range` : il permet de choisir le pas. Par exemple, pour répéter des instructions 10 fois mais en commençant par 10 jusqu'à 1, on peut écrire :

```python
for variable in range(10, 0, -1):
    print(variable)
```

## Les boucles while (tant que)
Les boucles `while` réutilisent une notion que l'on a vue dans le chapitre précédent : **les conditions**. En effet, dans ce contexte, on peut traduire le mot *while* par *tant que*. Il s'agira donc d'écrire des instructions qui auront comme logique:

*TANT QUE* **condition**, exécuter ces instructions :

    instruction1
    instruction2
    instruction3
    ...

Voici un exemple d'utilisation d'un boucle `while` permettant de continuer tant que l'utilisateur n'a pas saisi "stop" :

```python
choix = ""
while choix != "stop" :
    choix = input("Entrer 'stop' pour arrêter le programme > ")
```

Voici la syntaxe d'une boucle `while` qui se répète 5 fois :

```python
repetitions = 5
while repetitions > 0 :
    print("Je me répète")
    repetions = repetitions - 1
```

## Présentation du jeu
Crée un jeu où l'ordinateur choisi un nombre aléatoirement et le joueur doit essayer de le trouver.

- Le programme doit choisir un nombre aléatoire entre 1 et 10.
- Le joueur doit avoir 4 essais pour deviner le nombre.
- Une fois qu'une partie est terminée (gagnée ou perdue), le programme doit demander à l'utilisateur s'il souhaite rejouer et relancer une partie si celui-ci répond **"oui"**.
- À chaque essai, le programme doit :
    1. Demander au joueur de trouver le nombre situé entre 1 et 10.
    2. **Si** le nombre entré par le joueur est le bon, le jeu doit féliciter le joueur et arrêter la partie.
    3. **Sinon, si** le nombre entré n'est pas le bon et qu'il s'agissait du dernier essai, le jeu doit informer le joueur qu'il a perdu.
    4. **Sinon,** le jeu doit informer le joueur que son nombre est trop petit ou trop grand le cas échéant.

## Code du jeu
1. Importe le module `random`.
2. Crée une variable `nombre_ordi` qui contiendra un nombre aléatoire entre 1 et 10. (Voir l'exemple plus bas)
3. Crée une boucle `for` qui devra répéter 4 fois :
   1. **Demander** au joueur de trouver le nombre entre 1 et 10 et récupérer sa réponse dans une variable `nombre_joueur`.
   2. **Si** `nombre_joueur` = `nombre_ordi`, **afficher** *"Bravo, tu as trouvé le nombre !"*.
   3. **Sinon, si** `essai` = 3 (dernier essai donc), **afficher** *"Perdu, le nombre était {nombre_ordi}"* (en insérant la valeur de la variable).
   4. **Sinon, si** `nombre_joueur` < `nombre_ordi`, **afficher** *"Ton nombre est trop petit."*.
   5. **Sinon**, cela signifie que le nombre est trop grand, donc **afficher** *"Ton nombre est trop grand."*.
4. Crée une boucle `while` au début du programme (avant le choix du nombre aléatoire) qui devra se répéter indéfiniment.
5. À la fin de la boucle des essais :
   1. **Demande** à l'utilisateur s'il veut rejouer et stocke sa réponse dans une variable `reponse`
   2. **Si** reponse = "non", alors stopper la boucle à l'aide du mot-clef `break`.

## Générer un nombre aléatoire
Pour générer un nombre aléatoire, on utilise un module de python nommé `random`. Il dispose de plusieurs fonctions, notamment `random.randint()` pour générer un nombre entier aléatoirement. Voici comment l'utiliser :

```python
import random

nombre_ordi = random.randint(1, 10)
```

## Schéma logique du programme
![](./schema_nb_secret.png)

````python
import random
nombre_ordi = random.randint(1, 10)

for essai in range(0, 4):
    nombre_joueur = int(input("Trouve le nombre entre 1 et 10 : "))
    if nombre_joueur == nombre_ordi :
        print("Félicitations !")
        break
    elif essai == 3 :
        print(f"Perdu ! Le nombre était {nombre_ordi}")
    elif nombre_joueur < nombre_ordi :
        print("Ton nombre est trop petit.")
    else:
        print("Ton nombre est trop grand.")
````
